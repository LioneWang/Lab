**fmt(format string)**
- fmt = '[marker][line][color]'or'[color][line][marker]'

```python
#o:dot g:green
plt.plot([1, 2, 3], [1, 2, 3],'og', label='line 1')
#r:red s:square
plt.plot([1, 2, 3], [1, 4, 9], 'rs', label='line 2')
```

**class:Line2D**
- Method:
    - set_linewidth(int)

```python
# 2 class instances of Line2D
lines=plt.plot([1,2,3],[[1,2],[2,3],[4,5]])
lines[0].set_linewidth(5)
lines[1].set_alpha(0.2)
plt.show()
```
